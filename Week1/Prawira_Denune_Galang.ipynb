```python
import matplotlib.pyplot as plt
import seaborn as sns

display(df['Exam_Score'].describe())
display(df['Hours_Studied'].describe())
display(df['Distance_from_Home'].describe())

#Display
plt.figure(figsize=(8, 5))
sns.histplot(df['Exam_Score'], kde=True)
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df['Hours_Studied'], kde=True)
plt.title('Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df['Distance_from_Home'], kde=True)
plt.title('Distance from Home')
plt.xlabel('Distance_from_Home')
plt.ylabel('Frequency')
plt.show()



plt.figure(figsize=(8, 5))
sns.boxplot(x='Distance_from_Home', y='Exam_Score', data=df, order=['Near', 'Moderate', 'Far'])
plt.title('Exam Score by Distance from Home')
plt.xlabel('Distance from Home')
plt.ylabel('Exam Score')
plt.show()
display(df.groupby('Distance_from_Home')['Exam_Score'].describe())


plt.figure(figsize=(8, 5))
sns.boxplot(x='Motivation_Level', y='Exam_Score', data=df, order=['Low', 'Medium', 'High'])
plt.title('Exam Score by Motivation Level')
plt.xlabel('Motivation Level')
plt.ylabel('Exam Score')
plt.show()
display(df.groupby('Motivation_Level')['Exam_Score'].describe())

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Exam Score vs. Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
display(df.groupby('Hours_Studied')['Exam_Score'].describe())
```

we divide 44 hours of Hours_Studied to be 3 catagories
1. Low (1-14)
2. medium (15-29)
3. High (30-44)

```python
bins = [0, 13, 27, 44]
labels = ['1-13', '14-27', '28-44']

df['Hours_Studied_Range'] = pd.cut(df['Hours_Studied'], bins=bins, labels=labels, right=True, include_lowest=True)


mean_exam_by_hours_range = df.groupby('Hours_Studied_Range')['Exam_Score'].mean()


print("Average Exam Scores based on Range of Study Hours:")
print(mean_exam_by_hours_range)
```

```python
mean_exam_by_motivation = df.groupby('Motivation_Level')['Exam_Score'].mean()

print("Average Exam_Score base motivation level:")
print(mean_exam_by_motivation)
```

```python
mean_exam_by_distance = df.groupby('Distance_from_Home')['Exam_Score'].mean()

print("Average Exam Score base Distance from Home:")
print(mean_exam_by_distance)
```

**8. Which factors are positively correlated with exam results t**

```python
print("Descriptive Measures of Association with Exam_Score for ALL Variables:")
numerical_vars = df.select_dtypes(include=np.number).columns.tolist()
numerical_vars.remove('Exam_Score')

print("\nCorrelation Coefficients with Exam_Score (for Numerical Variables):")
for var in numerical_vars:
    correlation = df[var].corr(df['Exam_Score'])
    print(f"Correlation between {var} and Exam_Score: {correlation:.2f}")

print("\nMean Exam_Score across Categories (for Categorical Variables):")

categorical_vars = df.select_dtypes(include=['object', 'category']).columns.tolist()

for var in categorical_vars:
    print(f"\nMean Exam_Score by {var}:")
    mean_exam_by_category = df.groupby(var, dropna=False)['Exam_Score'].mean()
    print(mean_exam_by_category)
```

**9. Which variables have a distribution close to normal, explain your answer?**

```python
numerical_vars = df.select_dtypes(include=np.number).columns.tolist()
if 'Hours_Studied_Range' in numerical_vars:
    numerical_vars.remove('Hours_Studied_Range')

print("\nSkewness and Kurtosis for Numerical Variables:")
for var in numerical_vars:
    skewness = df[var].skew()
    kurtosis = df[var].kurtosis()
    print(f"{var}: Skewness={skewness:.2f}, Kurtosis={kurtosis:.2f}")

print("\nHistograms for Numerical Variables:")
for var in numerical_vars:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[var], kde=True)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()
```

-Hours_Studied: Skewness (0.01) and kurtosis (0.02) are very close to 0. The histogram also appears quite symmetrical and bell-shaped.

-Physical_Activity: Skewness (-0.03) and kurtosis (-0.06) are also very close to 0. The histogram appears relatively symmetrical, although somewhat uniform in shape across some values.

-Other variables, such as Attendance and Previous_Scores, have skewness close to 0, but their kurtosis is quite negative (-1.19), indicating a distribution that is flatter (platykurtic) than normal.

-Sleep_Hours has skewness close to 0 but negative kurtosis (-0.50).
